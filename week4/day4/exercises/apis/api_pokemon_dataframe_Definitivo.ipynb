```python
# En primer lugar, importo las librerías necesarias
import requests
import pandas as pd 
import numpy as np

#También la librería json para poder operar con la API
import json
```

```python
#Ahora me toca a través del enlace de la api, pasarlo a formato json

url = "https://pokeapi.co/api/v2/pokemon"

#Creando la variable api_pokemon, lo que estoy pidiendo es que se obtenga el archivo que obtiene ese enlace y que lo pase a formato json 
api_pokemon = requests.get(url).json()

#Si yo ahora hago un print de api_pokemon...

print(api_pokemon)
print('----------')
# Lo que obtengo es un diccionario, y así puedo acceder a los valores que yo desee
#Por ejemplo: al acceder a la clave ['count'] estoy obteniendo su valor, en este caso, 1118

print(api_pokemon['count'])
print('--------')
print(type(api_pokemon['count'])) #En este caso el valor de esta clave es de tipo entero

print('---------')
print(type(api_pokemon)) #La variable api_pokemon, como decíamos, es un diccionario

#Es muy importante conocer bien el json e incluso tomarse un tiempo para comprender los distintos tipos de objetos que hay y conocer las relaciones que tienen con las distintas librerías
```

```python
#Como ya sabemos, podemos ver las claves del diccionario a través de .keys()

print('Claves del diccionario api_pokemon:',api_pokemon.keys())
print('------------------------------------------------------------------------------------------------')

#Teniendo esto en cuenta, podremos crear un data frame en el que las claves serán las distintas columnas, y los valores se insertarán en cada fila

#Creamos el DataFrame de la siguiente forma:

df_api_pokemon = pd.DataFrame(api_pokemon)
df_api_pokemon
```
