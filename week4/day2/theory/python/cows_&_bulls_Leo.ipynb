
- [Cows_and_Bulls]

Say the number generated by the computer is 1038. An example interaction could look like this:

  Welcome to the Cows and Bulls Game! 
  
  >Enter a number:

  >> 1234

  >>> 2 cows, 0 bulls

  >> 1253

  >>> 1 cow, 1 bull
  
  ...
Until the user guesses the number or write "STOP"

```python
def cows_bulls(your_number):
    your_number = str(your_number)
    number = str(1038)
    cows = 0
    bull = 0
    for pos, value in enumerate(number):
        if number[pos] == your_number[pos] and value in your_number:
            cows +=1
        if value in your_number:

            bull += 1
    return f'{cows}, cows, {bull} bull'

cows_bulls(your_number=1234)
```

```python
#Después hemos introducido un while para repetirlo infinitas veces
password = '1223'
guess = 'x'

while guess != password and guess != 'STOP':
    cows = 0
    bulls = 0
    guess = input("Enter a number").upper()
    for pos, val in enumerate(guess):
        if val in password and val == password[pos]:
            cows += 1
        elif val in password:
            bulls += 1
    print(cows, "cows", bulls, "Bulls")
if cows == 4:
    print("Enhorabuena, has ganado el juego!")
else:
    print("Eres un perdedor nato!")
```
