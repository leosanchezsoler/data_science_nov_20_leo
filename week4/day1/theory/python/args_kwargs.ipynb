Parámetros *args y *kwargs en Python

```python
def mi_funcion():
    return 5

x = mi_funcion()
print(x)

```

```python
def mi_funcion(a):
    return a

x = mi_funcion(a=6)
print(x)
```

```python
#args es una tupla de los argumentos que doy a la función al invocarla
#Es obligatorio escribir un asterisco para args '*args'. El asterisco representa la indeterminación de la longitud.

#Todo lo que no tenga el nombre args

def mi_funcion(*args):
    """
    Suele describirse en la función lo que representa cada posición de *args
    """
    print(args)
    print(type(args))
    tres = args[1]
    print("tres:", tres)

mi_funcion(2,3)
```

```python
def suma(*args):
    x = args[0]
    y = args[1]
    print(x+y)

suma(4, 7)
```

```python
def suma_v2(*args):
    suma_total = 0
    for elem in args:
        suma_total = suma_total + elem
    return suma_total

x = suma_v2(2, 7, 9, 5, 10)
print(x)
```

```python
# Ejemplo *args

def args_example(*args):
    print("args:", args)
    print("type(args):", type(args))
    one = args[0]
    return one

one = args_example(7,8,9)
print("one:", one)
```

```python
*args = Tuple
** kwargs = Dict
```

```python
#kwargs se escribe con dos asteriscos '**kwargs'
#los valores de kwargs se muestran como un diccionario a través de keys y values 

def mi_function_kwargs(**kwargs):
    """
    **kwargs:
        -se especifican los tipos que hay 
    """
    print(kwargs)
    print(type(kwargs))
    x = kwargs["x"]
    print("x:", x)

mi_function_kwargs(x=2, a=6, b=8)
```

```python
def mi_function_kwargs(x, **kwargs):
    print(kwargs)
    print(type(kwargs))
    h = kwargs["b"]
    print("h:", h)

mi_function_kwargs(x=2, a=6, b=8)
```

```python
#args por defecto es una tupla vacía y kwargs por defecto es un diccionario vacío. En este caso, las dos funciones van a mostrarse sin error 

#Todo lo que no esté previamente definido como parámetro en una lista va a args si no está asignado y a kwargs si se le asigna una variable
def mi_function_kwargs(x, t, s=5, *args, **kwargs):
    print("x:", x)
    print("t:", t)
    print("s:", s)
    print("args:", args)
    print("kwargs:", kwargs)
    print(type(kwargs))
    h = kwargs["b"]
    print("h:", h)

mi_function_kwargs(2, 4, 6, "g", "p", a=6, b=8)
print("---------------")
mi_function_kwargs(3, 4, b=6)
```

```python
# Ejemplo **kwargs

def kwargs_example(**kwargs):
    print("kwargs:", kwargs)
    print("type(kwargs):", type(kwargs))
    one = kwargs["z"]
    return one

one = kwargs_example(x=1, y=4, z=7)
print("one:", one)
```
