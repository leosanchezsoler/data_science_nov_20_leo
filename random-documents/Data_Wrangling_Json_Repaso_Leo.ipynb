```python
import requests
import json
import pandas as pd
```

```python
r = requests.get(url='http://data.nba.net/prod/v2/2018/teams.json')
json_readed = json.loads(r.text)
```

```python
df_json = pd.DataFrame(json_readed)
df_json
```

```python
df2 = df_json.loc['standard':, 'league':'league'] #De este modo se accede como Dataframe a partir de una fila y una columna en concreto. Si hubiese más columnas a la derecha de league y dejase solo los dos puntos, se seleccionarúan todas a partir de ella
```
